```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace=True)
```

```python
df.shape
```

```python
# Encoding the categorical columns (Geography, Gender)
df = pd.get_dummies(df, columns=['Geography', 'Gender'], drop_first=True, dtype=int)
```

```python
df.head()
```

```python
X = df.drop(columns = ['Exited'])
y = df['Exited']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
y_train
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
# importing keras

import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(11, activation = 'relu', input_dim = 11))
model.add(Dense(11, activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
 model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train_scaled, y_train, epochs = 100, validation_split=0.2)
```

```python
model.layers[0].get_weights() # 121 weights + 11 bias = 132 trainable parameters of first layer
```

```python
model.layers[1].get_weights() # 121 weights + 11 bias = 132 trainable parameters of layer 2
```

```python
model.layers[2].get_weights() # 11 + 1 = 12 trainable_parameters
```

```python
# model.predict(X_test_scaled) # since the prediction is not in 1-0 format because we are using sigmoid function and sigmoid function gives probabilities between 1 and 0
```

```python
y_log = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_log>0.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
